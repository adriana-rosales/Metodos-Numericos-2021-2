```python
import numpy as np
import scipy as sp

def gaussSeidel(A,b,x,imax=100,tol=1e-8):
    L=np.tril(A)
    U=A-L
    Linv=np.linalg.inv(L)
    cumple=False
    k=0
    while (not cumple and k<imax):
        xk1=np.dot(Linv,b-np.dot(U,x))
        norma=np.linalg.norm(x-xk1)
        print('iteracion:{}->{} norma {}'.format(k,x,norma))
        cumple=norma<tol
        x=xk1.copy()
        k+=1

    if k<imax:
        return x
    else:
        raise ValueError  ('El sistema no converge')

def main():
    A = np.array([[-512,12,0,0,0],\
                [500,-512,12,0,0],\
                [0,500,-512,12,0],\
                [0,0,500,-512,12],\
                [0,0,0,-500,512]])
    b = np.array([0,0,0,0,9])
    x = np.array([0,0,0,0,0])
    x = gaussSeidel(A,b,x)
    print('Solución aproximada:',x)

if __name__ == "__main__": main() 
```

```python
import numpy as np
import scipy as sp

def gaussSeidel(A,b,x,imax=100,tol=1e-8):
    L=np.tril(A)
    U=A-L
    Linv=np.linalg.inv(L)
    cumple=False
    k=0
    while (not cumple and k<imax):
        xk1=np.dot(Linv,b-np.dot(U,x))
        norma=np.linalg.norm(x-xk1)
        print('iteracion:{}->{} norma {}'.format(k,x,norma))
        cumple=norma<tol
        x=xk1.copy()
        k+=1

    if k<imax:
        return x
    else:
        raise ValueError  ('El sistema no converge')

def main():                              #EDIT datos del sistema
    A=np.array([[3,-1,-1],\
                [-1,3,1],\
                [2,1,4]])
    b=np.array([1,3,7])
    x=np.array([0,0,0])
    x = gaussSeidel(A,b,x)
    print('Solución aproximada:',x)

if __name__ == "__main__": main() 

# Solución aproximada:

# x1=1
# x2=1
# x3=1
```
