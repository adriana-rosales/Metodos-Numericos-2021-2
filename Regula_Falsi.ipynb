#1: Libro

```python
#Import Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Definición de la función para el método de la Regla Falsa
def reglafalsa(f,x0,x1,tol=1e-5):                              #tol=
  '''
  ARGUMENTOS:

  Entrada
  -------
       f : Ecuación a resolver expresada en la forma f(x)=0
  x0, x1 : Puntos que encierran la raiz f(x0)*f(x1)<0
     tol : Tolerancia (valor por defecto 1e-5)
  Salida
  ------
      x : Raíz de la ecuación
  tabla : Tabla de iteraciones
  '''
  #validación del rango que encierra la raíz
  if f(x0)*f(x1)>0:
    raise Exception('Rango invalido, no cruza el eje x!!')
  x=x0 #valor inicial
  #encabezado de la tabla de iteracioens
  tabla=pd.DataFrame(columns=['x0','x','x1','f(x0)','f(x)','f(x1)'])
  #validar el criterio de convergencia
  while np.abs(f(x))>tol:
    x=(x0*f(x1)-x1*f(x0))/(f(x1)-f(x0)) #calcula x, método de Regla Falsa
    #inserta la iteración a la tabla
    tabla=tabla.append({'x0':x0,'x':x,'x1':x1,
                        'f(x0)':f(x0),'f(x)':f(x),'f(x1)':f(x1)},
                       ignore_index=True)
    #validar si la raíz se encuentra en el intervalo [x0,x]
    if f(x0)*f(x)<0:
      x1=x
    #validar si la raíz se encuentra en el intervalo [x,x1]
    else:
      x0=x
#Retornar la raíz y la tabla de iteraciones
  return x,tabla

#DEFINIR LA FUNCIÓN f(x)
def f(h):
  return np.pi*h**2*(30-h)/3-1000                             #EDIT f(x)

#CALL
h,t=reglafalsa(f,5,7.5)                                       #EDIT (,x0,x1,tol)
print('raíz aproximada=', h)
t
```

```python
#Import Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Definición de la función para el método de la Regla Falsa
def reglafalsa(f,x0,x1,tol=1e-4):                              #tol=

  #validación del rango que encierra la raíz
  if f(x0)*f(x1)>0:
    raise Exception('Rango inválido, no cruza el eje x!')
  x=x0 #valor inicial
  #encabezado de la tabla de iteracioens
  tabla=pd.DataFrame(columns=['x0','x','x1','f(x0)','f(x)','f(x1)'])
  #validar el criterio de convergencia
  while np.abs(f(x))>tol:
    x=(x0*f(x1)-x1*f(x0))/(f(x1)-f(x0)) #calcula x, método de Regla Falsa
    #inserta la iteración a la tabla
    tabla=tabla.append({'x0':x0,'x':x,'x1':x1,
                        'f(x0)':f(x0),'f(x)':f(x),'f(x1)':f(x1)},
                       ignore_index=True)
    #validar si la raíz se encuentra en el intervalo [x0,x]
    if f(x0)*f(x)<0:
      x1=x
    #validar si la raíz se encuentra en el intervalo [x,x1]
    else:
      x0=x
#Retornar la raíz y la tabla de iteraciones
  return x,tabla

#DEFINIR LA FUNCIÓN f(x)
def f(x):
  return x**3+2*x**2+10*x-20                                  #EDIT f(x)

#CALL
h,t=reglafalsa(f,1,2,tol=1e-04)                               #EDIT (,x₀=xI,x₁=xD,tol)
print('raíz aproximada x=', h)
t

# **  RESULTADOS "x=xM"  **
#con una Tol=10^-4 en iter=6
#La raíz aprox es:          x6=1.368757

# **COMPARAR SIGNOS**
# f(x=xM)=f(x1=xD) entonces xD=x M 
# f(x=xM)≠f(x1=xD) entonces xI=x M
```
