#1: cos(x), *x0=0*

1. `Función`
$$
f(x)=cos(x)
$$
2. `Punto inicial`
$$
x₀=0
$$
3. `Grado-orden`
$$
n-1=3
$$
4.   `Términos de la serie`
$$
n=4
$$

**Resultado:**
$$
f(x)=1-\frac{x^2}{2!}
$$

Fórmula general:
$$
f(x)=1-\frac{x^2}{2!}+\frac{x^4}{4!}-\frac{x^6}{6!}+\frac{x^8}{8!}-\frac{x^{10}}{10!}+\dots
$$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x0=0                                 #EDIT
n=4                                  #EDIT

x=sp.Symbol('x')
t=sp.cos(x).series(x,x0,n).removeO() #EDIT "sp."
print(t)
t=sp.lambdify(x,t,'numpy')

#g=plt.figure()

xx=np.linspace(x0-5,x0+5,1000)       #EDIT
yc=np.cos(xx)                        #EDIT "np."
yt=t(xx)

plt.plot(xx,yc,xx,yt)
plt.legend(['cos(x)','Serie de Taylor orden 4']) #EDIT NAME
plt.title('Serie de Taylor de $f(x)=cos(x)$')    #EDIT
plt.grid(True)

#g.savefig('seriet.pdf', bbox_inches='tight)
```

1. `Función`

f(x)=ln(x)
$$
2. `Punto inicial`
$$
x₀≠0
$$
$$
x₀=1
$$
3. `Grado-orden`
$$
n-1=4
$$

**Resultado:**

f(x)=(x-1)-\frac{1}{2}

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x0=1                                  #EDIT
n=5                                   #EDIT

x=sp.Symbol('x') 
t=sp.log(x).series(x,x0,n).removeO()  #EDIT "sp."
print(t)
t=sp.lambdify(x,t,'numpy')

#g=plt.figure()

xx=np.linspace(x0-0.9,x0+2,100)       #EDIT
yc=np.log(xx)                         #EDIT "np."
yt=t(xx)

plt.plot(xx,yc,xx,yt)
plt.legend(['ln(x)','Serie de Taylor orden 5']) #EDIT NAME
plt.title('Serie de Taylor de $f(x)=ln(x)$')    #EDIT
plt.grid(True)

#g.savefig('seriet.pdf', bbox_inches='tight)
```

#4: eˣ, *x0=0*

1. `Función`
$$
f(x)=e^x
$$
2. `Punto inicial`
$$
x₀=0
$$
3. `Grado-orden`
$$
n-1=4
$$
4.   `Términos de la serie`
$$
n=5
$$

**Resultado:**

$$
f(x)=1+1(x-0)+\frac{1}{2!}(x-0)^2+\frac{1}{3!}(x-0)^3+\frac{1}{4!}(x-0)^4+\dots
$$

Fórmula general:
$$
f(x)=\sum_{n=0}^{\infty}\frac{x^n}{n!}
$$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x0=0                                     #EDIT
n=5                                      #EDIT

x=sp.Symbol('x')
t=sp.exp(x).series(x,x0,n).removeO()   #EDIT "sp."
print(t)
t=sp.lambdify(x,t,'numpy')

#g=plt.figure()

xx=np.linspace(x0-4,x0+4,100)            #EDIT
yc=np.exp(xx)                            #EDIT "np."
yt=t(xx)

plt.plot(xx,yc,xx,yt)
plt.legend(['$e^x$','Serie de Taylor orden 5']) #EDIT NAME
plt.title('Serie de Taylor de $f(x)=e^x$')      #EDIT
plt.grid(True)

#g.savefig('seriet.pdf', bbox_inches='tight)
```
